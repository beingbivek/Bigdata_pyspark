```python
import os
os.environ["PYSPARK_PYTHON"] = "python"
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.appName("SparkSQL").master("local[*]").getOrCreate()
```

```python
sales_data = [

     (1,  "Laptop",     "Electronics", "2024-01-05", "North", "Alice", 2, 1200.0),

  (2,  "Phone",      "Electronics", "2024-01-12", "South", "Bob",   3,  800.0),

  (3,  "Tablet",     "Electronics", "2024-02-08", "East",  "Carol", 1,  600.0),

  (4,  "Headphones", "Accessories", "2024-02-15", "West",  "Dan",   4,  150.0),

  (5,  "Monitor",    "Electronics", "2024-03-01", "North", "Alice", 1,  400.0),

  (6,  "Keyboard",   "Accessories", "2024-03-14", "South", "Eve",   5,   90.0),

  (7,  "Laptop",     "Electronics", "2024-04-02", "East",  "Frank", 1, 1200.0),

  (8,  "Phone",      "Electronics", "2024-04-20", "West",  "Grace", 2,  800.0),

  (9,  "Mouse",      "Accessories", "2024-05-10", "North", "Bob",   6,   45.0),

  (10, "Tablet",     "Electronics", "2024-05-18", "South", "Carol", 3,  600.0),

  (11, "Monitor",    "Electronics", "2024-06-05", "East",  "Dan",   2,  400.0),

  (12, "Keyboard",   "Accessories", "2024-06-22", "West",  "Eve",   3,   90.0)

]
```

```python
schema = StructType([
    StructField("sale_id", IntegerType()),
    StructField("product", StringType()),
    StructField("category", StringType()),
    StructField("sale_date", StringType()),
    StructField("region", StringType()),
    StructField("customer", StringType()),
    StructField("qty", IntegerType()),
    StructField("unit_price", DoubleType())
])
```

```python
df = spark.createDataFrame(sales_data, schema = schema)
```

```python
df.show(5)
```

```python
df.printSchema()
```

```python
df = df.withColumn("revenue", col("qty") * col("unit_price"))
df.show(5)
```

## Implementing SPARK SQL

```python
df.createOrReplaceTempView("sales")
# creating a temporary view to run SQL queries on the DataFrame
```

```python
spark.sql("SHOW TABLES").show()
```

```python
query1 = """
SELECT * FROM sales
"""
result = spark.sql(query1)
result.show()
```

### Filter by category where category = Electronics

```python
query2 = """
SELECT * FROM sales
WHERE category = 'Electronics'
"""
result2 = spark.sql(query2)
result2.show()
```

### Group by product, Order by revenue

```python
query3 = """
SELECT product,
COUNT(*) AS orders,
SUM(qty) AS unit_sold,
ROUND(SUM(revenue), 2) AS total_revenue
FROM sales
GROUP BY product
ORDER BY total_revenue DESC
"""
result3 = spark.sql(query3)
result3.show()
```

### provide product, region ,qty, unit_price, revenue, revenue > 1000, category = Electronics, revenue DESC

```python
query4 = """
SELECT product, region, qty, unit_price, revenue
FROM sales
WHERE category = 'Electronics' AND revenue > 1000
ORDER BY revenue DESC
"""
result4 = spark.sql(query4)
result4.show()
```

### RANK products by revenue within each region
#### (aggrigate sql function)

```python
query5 = """
SELECT region, product, ROUND(SUM(revenue), 2) AS total_revenue,
RANK() OVER (PARTITION BY region ORDER BY SUM(revenue) DESC) AS revenue_rank
FROM sales
GROUP BY region, product
ORDER BY region, revenue_rank
"""
result5 = spark.sql(query5)
result5.show()
```

### subquery =  products over the average revenue

```python
query6 = """
SELECT product, category, total_revenue
FROM (
SELECT product, category, SUM(revenue) AS total_revenue
FROM sales
GROUP BY product, category
) prod_totals
WHERE total_revenue > (SELECT AVG(revenue) FROM sales)
ORDER BY total_revenue DESC
"""
result6 = spark.sql(query6)
result6.show()
```
